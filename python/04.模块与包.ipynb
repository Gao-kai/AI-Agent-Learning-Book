# 第四章 Python中的模块与包

### 1. 模块 Module

1. python中一个.py文件就默认是一个模块Module,模块中可以包含变量、函数、类等各种内容
2. python中的模块可以分为标准库模块（python内置）、自定义模块以及安装的第三方模块
3. 模块的名称要符合标识符规则，不要用标准库模块相同名称的模块名
4. 模块导入的方法有5种
    - import module 导入模块中所有内容
    - import module as 别名
    - from module import 具体内容1，具体内容2
    - from module import 具体内容1 as 别名1，具体内容2 as 别名2
    - from module import * 直接加载模块中所有内容到当前模块，可能引起命名冲突
5. 对于from module import * 这种导入方法可以在定义模块的时候通过__all__来限制可以供这种方法导入的模块内容
6. __name__在模块中是一个动态变量
    - 主程序启动，值为__main__（可以将测试模块内部方法的代码写在里面起到保护的效果）
    - 其他模块导入时，值为模块名称
7. python的标准库模块分为内置标准库模块(C语言编写)和非内置标准库模块（Py语言编写源代码存放在python的安装目录下,可以通过__file__访问）
    - copy模块
    - os模块
    - random模块
    - time模块
    - math模块
    - sys模块

### 2. 包 Package
1. python中将一个包含__init__.py文件的文件夹叫做包 Package
2. python中的包中可以包括多个子模块module
3. python中的包还可以包含子包


```python
# python中包的8种导入方法
import 包名.模块名
import 包名.模块名 as 模块别名
from 包名.模块名 import 类名, 函数名, 变量名
from 包名.模块名 import 类名 as 类别名1, 函数名 as 函数别名1, 变量名 as 变量别名1
from 包名.模块名 import *  # 导入模块中的所有内容 直接加载到当前作用空间

from 包名 import *
from 包名 import 模块名 
from 包名 import 模块名 as 模块别名
```

### 3. __init__.py文件

1. __init__.py文件是一个特殊的文件，它用于初始化包, 当导入包时, 会自动执行__init__.py文件中的代码,可以为空
2. __init__.py文件中可以包含包的元数据，例如包的版本号、作者、描述等
3. __init__.py文件中可以包含包的初始化逻辑
4. __init__.py文件中定义的变量、函数、类等，都会在from 包名 import *执行的时候全部加载到当前作用空间中，当然也可以通过__all__列表来指定要导出的变量、函数、类等
5. __init__.py文件中可以包含包的子模块、子包等
