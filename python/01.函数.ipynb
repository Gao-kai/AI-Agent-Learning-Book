## 2. 函数位置参数和关键字参数
1. 调用函数时传递的参数会按照形参定义时出现的顺序依次传递，这叫做位置参数
2. 调用函数时通过形参名=值的形式传递实参，这叫做关键字参数
    - 关键字参数的传递顺序不受形参限制
    - 关键字参数形参名和等号=中间最好不要加空格
3. 两者混用时，位置参数必须保证在关键字参数之前
4. 函数中限制位置参数和关键字参数
    - /之前只能传递位置参数
    - *之后只能传递关键字参数
    - /必须在*号之前，否则报错

```python
def person_info(name, /, age, *, gender, score):
    """
    打印用户基本信息
    :param name: 用户姓名
    :param age: 用户年龄
    :param gender: 用户性别
    :param score: 用户评价
    :return: None
    """
    print(f"用户的姓名是{name}")
    print(f"用户的年龄是{age}")
    print(f"用户的性别是{gender}")
    print(f"用户的评分是{score}")


person_info("Lily", age=18, gender="男", score=100)
person_info("Liam", 20, gender="女", score=99)
```

## 3. 默认参数和可选参数

1. 定义函数时通过形参名=值的方式可以为函数参数指定一个默认值，之后函数在调用的时候，这个参数可以传递也可以不传递，可以用关键字也可以用位置参数，并且默认参数一定在必选参数的后面。
2. 关键字可选参数用**kwargs来接受任意数量的关键字参数，kwargs中存放的数据类型是一个字典
3. 位置可选参数用*args来接受任意数量的位置参数，args中存放的数据类型是一个元组
4. 可选位置参数*args必须声明在可选关键字参数**kwargs前面
5. 函数在调用传递实参时也可以进行解包操作（等于js中展开运算符）
    - 位置参数解包 * ：用于将列表、元组等可迭代对象解包为位置参数。
    - 关键字参数解包 ** ：用于将字典解包为关键字参数，对应 JavaScript 的 {...obj} 对象展开：


```python
def person_info(name, *args, score=100, **kwargs):
    print(f"用户的姓名是{name}")
    print(f"用户的评分是{score}")
    print(f"可选位置参数是{args}")
    print(f"可选关键字参数是{kwargs}")


# 用户的姓名是Lily
# 用户的评分是100
# 可选位置参数是('QiangHua', 'BeiJing')
# 可选关键字参数是{'hobby': 'basketball', 'age': 18}
person_info("Lily", "QiangHua", "BeiJing", hobby="basketball", age=18)
```

## 6. 函数递归调用

关于递归调用一个最形象的例子就是把大事化简为一个个简单的小事来做，最后将小事的结果进行汇总。比如在一个100个人的长队伍中，你想知道自己的排位，只需要问前一个人，前一个人重复问前一个人，当问到第一个人的的时候得到一个确定的结果是1，然后依次后面每个人都加1得到最终结果。

```python
# 函数递归调用 先下潜后打印
# 执行栈：log(5) ->log(4) ->log(3) ->log(2) ->log(1)
# 下到最深一层log(1)执行打印操作
def log(times: int):
    if times > 1:
        log(times - 1)
    print("函数调用第{times}次")
```

```python
# 函数递归调用 先打印后下潜
# 执行栈：log(5) ->log(4) ->log(3) ->log(2) ->log(1)
# 先执行打印 然后下潜
def log(times: int):
    print("函数调用第{times}次")
    if times > 1:
        log(times - 1)
```

```python
# 使用递归计算某一个数的阶乘
# 阶乘就是所有小于等于自己的数的正整数的积
def mul(number: int):
    if number == 0:
        return 1
    return number * mul(number - 1)


print(mul(5))
print(mul(8))
```

## 7. 函数综合案例

```python
def calc_total(*nums):
    """
    计算运动量总和
    :param nums: 每一天的运动量
    :return: 总和
    """
    return sum(nums)


def calc_avg(sum, days=7):
    """
    计算平均运动量
    :param sum: 运动量总和
    :param days: 运动天数
    :return: 平均值
    """
    return sum / days


def check_success(total, goal=120):
    """
    检测挑战是否成功
    :param total: 实际总和
    :param goal: 目标
    :return:
    """
    if total >= goal:
        return "✅挑战成功"
    else:
        return "❌挑战失败"


def main(title, days):
    nums = []
    for index in range(1, days + 1):
        nums.append(int(input(f"第{index}天运动记录：")))
    total = calc_total(*nums)
    result = check_success(total)
    print(result)


main(title="俯卧撑", days=7)
```

## 二、函数高级特性

### 1. 函数对象

1. python中的函数也是一个对象
2. 函数既然是对象，那么也可以动态给这个函数添加属性
3. 函数也可以赋值给其他变量，等于是变量和函数名都指向该函数对象的堆内存地址
4. 函数的不可变参数（字符串、数字、布尔值）时，在函数运行时修改不可变参数的值，不会影响到原参数，这是因为修改不可变参数的值会重新在内存中分配一个新的空间来存储这个新的值，而原参数原来指向的内存地址不会改变，因此值就不会发生变化
5. 函数接受可变参数（列表、元组、字典、实例对象）时，在函数运行时修改可变参数的成员，由于是引用传递，所以会直接修改原对象
6. 函数可以当作参数也可以当作返回值

### 2. 函数多返回值和参数的打包与解包

1. 函数可以返回多个值，默认用元组包裹，也可以通过列表形式返回，并且返回值都可以直接通过解构拿到值
2. 函数形参打包接受参数
   - *args：打包接受多个位置参数，将所有位置参数打包成一个元组
   - **kwargs：打包接受多个关键字参数，将所有关键字参数打包成一个字典
3. 函数调用实参传递参数
   - *变量名 将元组中的元素依次解包传递给函数
   - **变量名 将字典中的键值对解包传递给函数

```python
# 位置参数必须放在关键字参数前面
def get_user_info(id,*args,gender,**kwargs):
    print(id)
    print(args)
    print(kwargs)
    print(args)

    return id,gender,args,kwargs

hobby = ["篮球","足球","跑步"]
school = "清华大学"
score = 90
name = "张三"
age = 18

# 元组
user_info_tuple = (score,name,age)

# 字典
user_info_dict = {
    "hobby": hobby,
    "school": school,
}

# 将元组参数解包成一个个独立的位置参数后传递给函数，在函数中通过*args接收
# 将字典参数解包成一个key value形式的关键字参数后传递给函数，在函数中通过**kwargs接收
get_user_info(1001,gender="男",*user_info_tuple,**user_info_dict)
```
