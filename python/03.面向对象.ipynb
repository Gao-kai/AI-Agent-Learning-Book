# 第三章 Python 面向对象
### 1. 类的定义

1. 类的初始化方法 __init__
2. 类的属性和方法
3. 类的工厂函数
4. 类的静态方法
5. 实例方法


```python
from datetime import datetime


class Person:
    """
    定义类属性
    1. 类属性是类的所有实例都共享的属性，用于存储类的属性值
    2. 类属性可以直接被类访问，也可以被实例访问（查找路径：实例自身->类）
    """
    max_age = 150
    planet = "地球"

    """
    初始化方法 __init__
    方法用于在实例化一个类的时候自动调用，类似js中的constructor方法
    此方法中self表示当前实例化对象的引用，用于访问对象的属性和方法
    """

    def __init__(self, name, age, gender):
        """
        实例属性
        1. 调用__init__方法时，会自动创建实例属性，用于存储实例的属性值
        2. 实例创建完成之后也可以在实例上添加新的属性
        3. 实例属性只可以供实例调用，类无法直接访问实例属性
        """
        self.name = name
        self.gender = gender
        if int(age) > Person.max_age:
            print("年龄不能超过最大年龄")
            self.age = Person.max_age
        else:
            self.age = int(age)

    """
    实例方法 say_hello、run
    所有通过Person类创建出来的实例都共享此方法，主要是供实例进行调用
    因此实例方法的调用是从实例->类一层层的进行调用的
    """

    def say_hello(self):
        print(f"你好，我是{self.name}，我今年{self.age}岁，我是{self.gender}")

    def run(self, distance):
        print(f"{self.name}正在跑步，距离是{distance}米")

    """
    类方法 sleep
    1. 通过装饰器@classmethod可以将方法定义为类方法
    2. 类方法接收的参数是当前类本身以及若干自定义的参数，由于接收了cls参数，因此类上的属性和其他方法都可以直接访问
    3. 类方法通常实现与类有关的逻辑,比如修改类属性、创建类的工厂方法
    4. 类方法是通过类进行调用的，比如Person.sleep(8)
    5. 不建议通过实例去调用类的方法，最好是类调用类的方法，实例调用实例方法
    """

    @classmethod
    def sleep(cls, hours):
        print(f"{cls.planet}上的人正在睡觉，时间是{hours}小时")

    """
    类方法 create
    1. 创建类的实例可以通过直接调用类Person初始化方法init传入参数，实现创建实例对象的功能
    2. 也可以通过直接调用类的工厂方法create传入参数，实现创建实例对象的功能（参数更加灵活）
    """

    @classmethod
    def create(cls, person_info):
        name, age, gender = person_info.split("-")
        current_year = datetime.now().year
        age = current_year - int(age)
        return cls(name, age, gender)

    """
    静态方法 is_adult、mack_id_card
    1. 静态方法是不依赖于类或实例的，直接通过类名调用
    2. 静态方法接收的参数是若干自定义的参数，由于没有接收cls参数，因此类上的属性和其他方法不能直接访问
    3. 静态方法通常实现类的工具方法
    4. 不建议通过实例去调用类的方法，最好是类调用类的方法，实例调用实例方法
    """

    @staticmethod
    def is_adult(year):
        current_year = datetime.now().year
        age = current_year - year
        return age >= 18

    @staticmethod
    def mack_id_card(id_number):
        if len(id_number) != 18:
            print("身份证号长度错误")
            return False
        return id_number[:6] + "********" + id_number[-4:]


"""
实例化一个类
1. 通过实例上的__dict__属性可以查看实例的所有属性和方法
2. 通过实例上的__class__属性可以查看实例的类
3. 通过type方法可以查询当前实例是由哪一个类名创建出来的
"""
p1 = Person("张三", 18, "男")
p2 = Person("李四", 190, "女")

print(p1.__dict__)  # {'name': '张三', 'age': 18, 'gender': '男'}
print(p1.__class__)  # <class '__main__.Person'>
print(type(p1))  # <class '__main__.Person'>

p1.say_hello()  # 你好，我是张三，我今年18岁，我是男
p2.say_hello()  # 你好，我是李四，我今年19岁，我是女

"""
类的工厂函数创建实例对象
"""
p3 = Person.create("王五-1997-男")
p3.say_hello()

print(Person.is_adult(2008))
print(Person.mack_id_card("44030419900101001X"))
```

### 2. 类的继承

```python
"""
类的继承
"""


class Student(Person):
    """
    子类的初始化可以借助super()方法调用父类的初始化方法，也可以在子类的初始化方法中添加自己的属性
    """

    def __init__(self, name, age, gender, class_name):
        super().__init__(name, age, gender)
        self.class_name = class_name

    """
    子类的方法重写
    重写后子类在调用时会优先调用子类的方法，而不是父类的方法
    """

    def say_hello(self):
        print(f"你好，我是{self.name}，我今年{self.age}岁，我是{self.gender}，我的班级是{self.class_name}")
        # 调用父类的方法
        super().say_hello()


s1 = Student("张三", 18, "男", "1班")
s1.say_hello()
print(s1.__dict__)
```

### 4. 类的多重继承

```python
class Worker:
    def __init__(self, job, salary):
        self.job = job
        self.salary = salary

    def say(self):
        print(f"我是{self.job},我有{self.salary}元")


class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def say(self):
        print(f"我是{self.name},我今年{self.age}岁,我是{self.gender}")


"""
多重继承：Student类可以一次性继承Worker类和Person类
注意：多重继承时，super()方法返回的是第一个继承的类
"""


class Student(Worker, Person):
    def __init__(self, class_name, score):
        Worker.__init__(self, "学生", 1000)
        Person.__init__(self, "张三", 18, "男")
        self.class_name = class_name
        self.score = score

    def say(self):
        print(f"我是{self.class_name},我有{self.score}分")


# 查看Student类的mro属性，即方法解析顺序,可以查看实例调用的流程 实例自身 -> 父类Worker -> 父类Person -> object类
# 注意：mro属性是一个元组，元组中的元素是类的名称
# (<class '__main__.Student'>, <class '__main__.Worker'>, <class '__main__.Person'>, <class 'object'>)
print(Student.__mro__)
```

### 5. 类的访问权限和getter setter方法
1. 加了__前缀的属性比如__age，只能在类内部访问
2. 加了_前缀的属性比如_name，只能在类内部调用和子类中调用，不能在类外部调用
3. 普通属性可以在类内部、类外部以及子类中都可以访问
4. 注册getter setter方法可以对属性的访问进行限制，比如判断是否是18位身份证号，是否是18岁以上的人
5. 注册getter setter方法可以对属性的访问进行限制，比如判断是否是18岁以上的人

```python
class Person:
    def __init__(self, name, age, id_card):
        self.name = name  # 公有属性
        self._age = age  # 保护属性
        self.__id_card = id_card  # 私有属性

    # 注册id_card属性的getter方法
    # 1. 外部实例也可以访问私有属性的值，但是可以加一些判断逻辑,比如加密后返回给用户
    @property
    def id_card(self):
        return self.__id_card[:6] + "********" + self.__id_card[-4:]

    # 注册id_card属性的setter方法
    # 1. 外部实例也可以设置私有属性的值，但是可以加一些判断逻辑,比如判断是否是18位身份证号
    @id_card.setter
    def id_card(self, value):
        if len(value) != 18 or not value.isdigit():
            raise ValueError("身份证号格式错误")
        self.__id_card = value


p1 = Person("张三", 18, "44030419900101001X")
print(p1.id_card)
p1.id_card = "440304199001010019"
print(p1.id_card)
```

### 6. 类的魔法方法

我们把以__开头和结尾的方法称为魔法方法，也称为特殊方法。
它的特点是不需要手动调用，而是由Python解释器在满足特定条件时自动调用。

1. __init__方法：构造方法，用于初始化对象的属性
2. __str__方法：当调用print(obj)或者str(obj)时，会自动调用该方法，返回对象的字符串表示
3. __len__方法：当调用len(obj)时，会自动调用该方法，返回对象的长度
4. __lt__方法：当调用obj < other时，会自动调用该方法，返回一个布尔值，表示obj是否小于other对象
5. __gt__方法：当调用obj > other时，会自动调用该方法，返回一个布尔值，表示obj是否大于other对象
6. __eq__方法：当调用obj == other时，会自动调用该方法，返回一个布尔值，表示obj是否等于other对象
7. __getattr__方法：当访问obj上不存在的属性时，会自动调用该方法


```python
class Person:
    def __init__(self, name, age, id_card):
        self.name = name  # 公有属性
        self._age = age  # 保护属性
        self.__id_card = id_card  # 私有属性

    def __str__(self):
        return f"姓名:{self.name},年龄:{self._age},身份证号:{self.__id_card}"

    def __getattr__(self, name):
        raise AttributeError(f"对象{self.__class__.__name__}没有属性{name}")

    def __lt__(self, other):
        return self._age < other._age

    def __gt__(self, other):
        return self._age > other._age

    def __eq__(self, other):
        return self._age == other._age

    def __len__(self):
        return len(self.__dict__)


p1 = Person("张三", 18, "44030419900101001X")
p2 = Person("李四", 20, "44030419900101002X")
# 触发__str__方法
print(p1)
print(p2)

# 触发__lt__方法
print(p1 < p2)
# 触发__gt__方法
print(p1 > p2)
# 触发__eq__方法
print(p1 == p2)
# 触发__len__方法
print(len(p1))
# 触发__getattr__方法
print(p1.address)
```

### 8. 标准多态和鸭子多态

1. 标准多态：同一个类的方法在调用时，根据不同的对象，可能会有不同的行为
   - 条件1:传入的对象必须是类的实例或者子类的实例
   - 条件2:各个类内部必须对同一方法进行重写，才可能呈现不同行为
   - 条件3:类之间必须具有继承关系

2. 鸭子多态：鸭子类型是一种编程风格，它不检查对象的类型，只关注对象是否可以调用某个方法，如果可以调用，就认为是鸭子类型

### 9. 抽象类

1. Python中的抽象类不能被直接实例化，它通常作为其他类的规范
2. 抽象类中可以包含抽象方法，也可以包含非抽象方法
3. 当其他子类继承抽象类之后必须实现抽象方法
4. 一个类通过继承ABC类是实现抽象类的定义
5. 一个方法通过@abstractmethod装饰器来定义为抽象方法

```python
from abc import ABC, abstractmethod


# 定义抽象类AbstractPerson，包含抽象方法whoami
class AbstractPerson(ABC):
    @abstractmethod
    def whoami(self):
        pass

    # 也可以定义非抽象方法
    def get_name(self):
        print(f"获取姓名是:{self.name}")


# 继承抽象类AbstractPerson
class Banker(AbstractPerson):
    def __init__(self, name):
        self.name = name

    # 必须实现抽象方法whoami
    def whoami(self):
        print("我是银行的柜员")


b = Banker("张三")
b.whoami()
b.get_name()
```

### 10. 练习

```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender


from datetime import datetime


class Student(Person):
    def __init__(self, name, age, gender, id, scores):
        super().__init__(name, age, gender)
        self.id = f"{datetime.now().strftime('%Y%m%d%H%M%S')}{id:03d}"
        self.scores = scores or {}

    # 魔法方法
    def __str__(self):
        return f"姓名:{self.name},年龄:{self.age},性别:{self.gender},学号:{self.id},成绩:{self.scores}"

    # 计算平均值
    def get_average(self):
        if not self.scores:
            return 0
        return f"{sum(self.scores.values()) / len(self.scores):.1f}"

    # 添加成绩
    def add_score(self, subject, score):
        self.scores[subject] = score


class Manager:
    id = 0

    def __init__(self):
        self.students = []

    def run(self):
        while True:
            choice = int(input("请输入您的选择：\n1.添加学生\n2.删除学生\n3.查看所有学生\n4.设置成绩\n5.退出\n"))
            if choice == 1:
                self.add_student()
            elif choice == 2:
                self.delete_student()
            elif choice == 3:
                self.view_all_students()
            elif choice == 4:
                self.set_score()
            elif choice == 5:
                break
            else:
                print("输入错误，请重新输入")

    # 添加学生
    def add_student(self):
        name = input("请输入学生姓名")
        age = int(input("请输入学生年龄"))
        gender = input("请输入学生性别（男/女）")
        student = Student(name, age, gender, Manager.id + 1, {})
        Manager.id += 1
        self.students.append(student)
        print(f"添加成功，学生的学号是:{student.id}")

    # 删除学生
    def delete_student(self):
        id = input("请输入学生学号")
        target_student = None
        for student in self.students:
            if student.id == id:
                print("学号查询成功")
                target_student = student

        if target_student:
            self.students.remove(target_student)
            print(f"删除成功，学生的学号是:{target_student.id}")
        else:
            print("删除失败，学号不存在")

    # 查看所有学生
    def view_all_students(self):
        if self.students:
            for student in self.students:
                print(student)
        else:
            print("当前没有学生信息")

    # 设置成绩
    def set_score(self):
        id = input("请输入学生学号")
        if self.students:
            for s in self.students:
                if s.id == id:
                    print("学号查询成功")
                    score_str = input("请输入成绩，格式为：科目-成绩，多个成绩用逗号隔开")
                    score_list = score_str.replace("，", ',').split(",")

                    for item in score_list:
                        subject, score = item.split('-')
                        subject = subject.strip()
                        score = float(score.strip())
                        s.add_score(subject, score)
                    print("成绩设置成功")
                    return

            print("学号查询失败，请重新输入")

        else:
            print("当前没有学生信息")


m = Manager()
m.run()
```

### 10. Python中的内存结构

1. 堆内存
   python中的对象都保存在堆内存中，每一个对象都有一个唯一的内存地址
   对于不可变对象来说，重新赋值会重新创建一个对象，分配新的内存地址，原本的指针会指向新的对象，而旧的对象会被垃圾回收机制回收

   对于可变对象来说，重新赋值不会重新创建一个对象，而是直接修改对象的内容，而指针不会改变

2. 栈内存
   python中的变量都保存在栈内存中，然后通过内存地址来访问堆内存中的数据

3. 可变对象
   可变对象是指在对象创建后可以被修改的对象，例如列表、字典、集合以及所有自定义类的实例对象
4. 不可变对象
   不可变对象是指在对象创建后不能被修改的对象，例如字符串、元组、整数、布尔、None等
